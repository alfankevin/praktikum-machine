# Segmentasi Gambar dengan Clustering

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow # path cv2.imshow di google colab
```

## Citra 1

```python
# Load Gambar
img1 = cv2.imread('/content/drive/MyDrive/plat/plat1.jpg')

# Tampilkan gambar
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(img1);

h, w, ch = img1.shape
data = img1 / 255.0
data = data.reshape(h * w, ch)
data.shape
```

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # choose a random subset
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(8, 4))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=14)

plot_pixels(data, title='Input color space: 16 million possible colors')


from sklearn.cluster import KMeans

# Melakukan segmentasi pada gambar dengan K-Means
k = 2  # Jumlah cluster yang diinginkan
kmeans = KMeans(n_clusters=k)
kmeans.fit(data)

# Mendapatkan label cluster untuk setiap pixel
labels = kmeans.labels_

# Reshape label menjadi bentuk asli gambar
segmented_img1 = labels.reshape(h, w)

# Menampilkan perbandingan gambar sebelum dan sesudah segmentasi
plt.figure(figsize=(16, 6))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title('Citra Asli')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(segmented_img1, cmap='gray')
plt.title('Hasil Segmentasi')
plt.axis('off')

plt.show()
```
